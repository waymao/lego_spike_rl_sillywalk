```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def compute_cum_result(epn):
    mat = np.zeros(len(epn))
    sum = 0
    for idx, item in enumerate(epn):
        sum += item
        mat[idx] = sum
    return mat
```

## Q-Learning

```python
from data.qlearning import qlearning_collection
```

```python
qlearning_collection = np.array(qlearning_collection)
qlearning_rewards = qlearning_collection[:, :, :, 0]
qlearning_rewards.shape
qlearning_avg = np.average(qlearning_rewards, axis=0)
qlearning_avg_cum = np.cumsum(qlearning_avg, axis=1)
qlearning_avg_cum
```

```python
fig, axs = plt.subplots()
for item, epn in enumerate(qlearning_avg_cum):
    axs.plot(epn, label=("Q-Learning episode " + str(item)))
fig.set_dpi(100)
fig.legend()
axs.set_xlabel("Timestep")
axs.set_ylabel("Cumulative reward")
fig.savefig("qlearning_result.pdf")
```
